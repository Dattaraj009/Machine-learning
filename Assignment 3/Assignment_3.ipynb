```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load data from the CSV file
file_path = 'D:\\advertising.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data.head()
```

```python
# Scatter plot: TV vs Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TV', y='Sales', data=data)
plt.title('Scatter Plot: TV Advertising vs Sales')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.show()
```

```python
avg_sales = data[['TV', 'Radio', 'Newspaper', 'Sales']].mean()
plt.figure(figsize=(10, 6))
avg_sales.plot(kind='bar', color=['blue', 'green', 'red', 'purple'])
plt.title('Average Sales per Advertising Medium')
plt.xlabel('Advertising Medium')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Box plot: Sales distribution by Advertising Medium
plt.figure(figsize=(5, 6))
sns.boxplot(data=data[['TV', 'Radio', 'Newspaper', 'Sales']])
plt.title('Box Plot: Sales Distribution by Advertising Medium')
plt.xlabel('Advertising Medium')
plt.ylabel('Sales')
plt.show()
```

```python
# Pie chart: Proportion of Total Advertising Budget
total_budget = data[['TV', 'Radio', 'Newspaper']].sum()
plt.figure(figsize=(6, 10))
plt.pie(total_budget, labels=total_budget.index, autopct='%1.1f%%', colors=['blue', 'green', 'red'])
plt.title('Proportion of Total Advertising Budget')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D line
ax.plot(data['TV'], data['Radio'], data['Newspaper'], label='Advertising Budgets', marker='o')

# Add labels and title
ax.set_xlabel('TV Budget')
ax.set_ylabel('Radio Budget')
ax.set_zlabel('Newspaper Budget')
ax.set_title('3D Line Chart: Sales over Different Advertising Budgets')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
